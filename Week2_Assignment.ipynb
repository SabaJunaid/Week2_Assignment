## ASSIGNMENT 1: The Data Analyzer

1. Write a function get_numbers() that asks the user to enter numbers separated by spaces and returns them as a list of floats.


```python
def get_numbers():
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(num) for num in user_input.split()]
    return numbers
data = get_numbers()
print(data)
```

2. Write a function calculate_mean(data) that returns the mean (average).

```python
def calculate_mean(data):
    total = sum(data)
    count = len(data)
    mean = total / count
    return mean

data = [10.0, 12.0, 23.0, 23.0, 16.0, 23.0, 21.0, 16.0]
mean_value = calculate_mean(data)
print("Mean:", mean_value)
```

3. Write a function calculate_variance(data, mean) that calculates sample variance.

```python
def calculate_variance(data, mean):
    n = len(data)
    variance = sum((x - mean) ** 2 for x in data) / (n - 1)  # sample variance
    return variance

data = [10.0, 12.0, 23.0, 23.0, 16.0, 23.0, 21.0, 16.0]
mean_value = 18.0
variance_value = calculate_variance(data, mean_value)
print("Variance:", round(variance_value, 2))
```

4. Write a function calculate_std_dev(variance) that returns the square root of the variance.

```python
import math

def calculate_std_dev(variance):
    return math.sqrt(variance)

variance_value = 24.57
std_dev_value = calculate_std_dev(variance_value)
print("Standard Deviation:", round(std_dev_value, 2))
```

5. In the main program:
   - Call get_numbers()
   - Call calculate_mean()
   - Call calculate_variance()
   - Call calculate_std_dev()
   - Print results in formatted output.

```python
import math

# Step 1: Function to get numbers
def get_numbers():
    return [float(x) for x in input("Enter numbers separated by spaces: ").split()]

# Step 2: Function to calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# Step 3: Function to calculate variance
def calculate_variance(data, mean):
    n = len(data)
    return sum((x - mean) ** 2 for x in data) / (n - 1)  # sample variance

# Step 4: Function to calculate standard deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)

# --- MAIN PROGRAM ---
# Call get_numbers()
data = get_numbers()

# Call calculate_mean()
mean_value = calculate_mean(data)

# Call calculate_variance()
variance_value = calculate_variance(data, mean_value)

# Call calculate_std_dev()
std_dev_value = calculate_std_dev(variance_value)

# Print results
print("\n--- Statistical Analysis ---")
print("Data:", data)
print("Mean:", round(mean_value, 2))
print("Variance:", round(variance_value, 2))
print("Standard Deviation:", round(std_dev_value, 2))
print("---------------------------")
```

## ASSIGNMENT 2: The Student Class System

1. Create a base class Person with attributes name and age.


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

2. Create method display_info() that prints name and age.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
```

3. Create derived class Student that inherits Person.

```python
class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)  # call parent constructor
        self.student_id = student_id
        self.major = major
```

5. Override display_info() to print student details.

```python
class Student(Person):  # Question 5
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)  # inherit name and age
        self.student_id = student_id
        self.major = major

    def display_info(self):  # override display_info
        super().display_info()  # call Person's display_info
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")
```

```python
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)  # inherit from Student
        self.thesis_topic = thesis_topic  # new attribute for graduate student
```

```python
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    def display_info(self):
        super().display_info()  # print name, age, student ID, major
        print(f"Thesis Topic: {self.thesis_topic}")
```

3. Implement area() returning width * height.

4. Overload + operator using __add__ to return new rectangle.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    def area(self):
        return self.width * self.height

    # Overload + operator
    def __add__(self, other):
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)
```

5. Overload == operator using __eq__ to compare areas.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    def area(self):
        return self.width * self.height

    # Overload + operator
    def __add__(self, other):
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)

    # Overload == operator
    def __eq__(self, other):
        return self.area() == other.area()
```

6. Create rectangles rect1(4,5) and rect2(2,3).

```python
# Create rectangle objects
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)

# Print rectangles
print("Rectangle 1:", rect1)
print("Rectangle 2:", rect2)
```

7. Print rectangles and areas.

```python
# Print areas
print("Area of rect1:", rect1.area())
print("Area of rect2:", rect2.area())
```

```python
# Add rectangles using overloaded +
rect3 = rect1 + rect2

# Print the new rectangle
print("\nrect3 (rect1 + rect2):", rect3)
```
